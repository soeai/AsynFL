```python
import pickle
import numpy as np
```

```python
def load_data(path):
    with open(path, "rb") as f:
        dataset = pickle.load(f)
    X = dataset[:, :-1]
    y = dataset[:, -1]
    return X, y
```

```python
data_folder = "./ember_data/chunk_1.pickle"
X_chunk, y_chunk = load_data(data_folder)
```
